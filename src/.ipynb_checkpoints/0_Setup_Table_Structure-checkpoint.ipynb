```python
# TABLE 2 races columns; 
# regatta
# race
# raceShort
# raceId
# maxWindSpd_kts
# minWindSpd_kts
# avgWindDir_deg
# firstlegbearing_deg

# coming from races, windsummary and firstlegbearing
filenames = ['races', 'windsummary', 'firstlegbearing']

for serv in eventSelection:
    d = download.download(server = serv['name'], regattasContaining = serv['regattaNameContaining'],\
     racesContaining = serv['raceNameContaining'])
#     regattas = d.getRegattas()

    for regatta in regattas:
#         races = d.getRaces(regatta['name'])
        directory_regatta =  parentdir + '/' + outdirbase + webAddressToFilename(serv['name']) + webAddressToFilename(regattaLoc(regatta['name']))
        filepath = directory_regatta + 'races.json'
        with open(filepath) as json_file:
            print('file extracted: ', filepath)
            data = json.load(json_file)
        dic_data = {'regatta': regatta['name'], 'data': data}
        races = cd.transformJsonToPD(dic_data, 'races')
        
        
        filepath = directory_regatta + 'windsummary.json'
        with open(filepath) as json_file:
            print('file extracted: ', filepath)
            data = json.load(json_file)
        dic_data = {'regatta': regatta['name'], 'data': data}
        windsummary = cd.transformJsonToPD(dic_data, 'windsummary').set_index(['regatta', 'raceShort'])   
        
        flb = []
        for index, race in races.iterrows():
            directory_race = parentdir + '/' + outdirbase +serv['name'] + webAddressToFilename(raceLoc(race['regatta'], race['race'])) ##TODO niet zo beun
            filepath = directory_race + 'firstlegbearing.json'
            with open(filepath) as json_file:
                print('file extracted: ', filepath)
                data = json.load(json_file)
            dic_data = {'regatta': regatta['name'], 'race': race['race'], 'data': data}
            firstlegbearing = cd.transformJsonToPD(dic_data, 'firstlegbearing') 
            flb.append(firstlegbearing)
        flb = pd.concat(flb).set_index(['regatta', 'race'])
        
    
        races = races.set_index(['regatta','raceShort'])
        result = pd.concat([races, windsummary], axis=1, join='outer').reset_index().set_index(['regatta', 'race'])
        result = pd.concat([result, flb], axis=1, join='outer').reset_index()
        
        print('with as result', result)
        result.to_sql('races', con = engine, index = False, if_exists = 'append')
        


```

```python
# UPDATE leg_nr in positions
regattass = [" 'Tokyo 2019 - 49er FX' ",
            " 'Toyko 2019 - 470 Women'",
            " 'Tokyo 2019 - 470 Men'",
            " 'Tokyo 2019 - RS:X Men'",
            " 'Tokyo 2019 - Finn'",]
for reg in regattass:
    print('working on ', reg)
    cursor.execute("""
    UPDATE positions
    SET    positions.leg_nr = test_markpassings.leg_nr
    FROM   regattas 
    INNER JOIN races on races.regatta = regattas.regatta
    INNER JOIN race_comp on race_comp.regatta = races.regatta
        and race_comp.race = races.race
    INNER JOIN comp_leg on comp_leg.regatta = race_comp.regatta
            and comp_leg.race = race_comp.race
            and comp_leg.comp_id = race_comp.comp_id
    INNER JOIN positions on  positions.regatta = race_comp.regatta
            and positions.race = race_comp.race
            and positions.comp_id = race_comp.comp_id
    INNER JOIN test_markpassings on test_markpassings.regatta = comp_leg.regatta
        and test_markpassings.race = comp_leg.race
        and test_markpassings.comp_id = comp_leg.comp_id
        and test_markpassings.leg_nr = comp_leg.leg_nr
    WHERE positions.timepoint_ms >= test_markpassings.begin_leg_ms 
        and positions.timepoint_ms < test_markpassings.end_leg_ms	
        and regattas.regatta =  """ + reg
        )
    print('finished with ', reg)
    cnxn.commit()
```
